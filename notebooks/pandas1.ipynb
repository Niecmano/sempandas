### U pandasu se biranje podataka iz dataseta moze izvršiti primenom različitih funkcija nad njim. Možemo npr. koristiti funkcije `loc` i `iloc`:

### - Kada se selekcija redova i kolona vrši na osnovu njihovog rednog broja, trebalo bi koristiti `iloc`

### - Kada se izbor zasniva na uslovu koje redovi treba da zadovolje i/ili kolone treba izabrati na osnovu njihovih naziva, treba koristiti `loc` 

```python
df.iloc[4:12,:4]
```

### Dakle, `iloc` prima uređeni par: [indeksi redova koji će biti uključeni u rezultat, indeksi kolona koje će biti u rezultatu]. Pogledajmo i `loc` koji radi sa istim uređenim parom, samo umesto indeksa navode se logički izrazi i liste željenih kolona.

```python
df.loc[df['Year']==2023,['Year','Industry_name_NZSIOC','Units','Value']]
```

### Umesto importovanja gotovog dataseta iz csv, json i sl. fajlova, pandas nudi mogućnost ručnog kreiranja datafrejma, navođenjem naziva svake kolone i njihovih vrednosti. Nakon toga moguće je i indeksirati svaki red preko defaultnog parametra index.

```python
df1 = pd.DataFrame({'grad': ['BG', 'NS', 'NI', 'KG'],
                    'glavni': ['DA', 'NE', 'NE', 'NE'],
                    'popul': ['1 mil', '300k', '200k', '150k']},
                   index=[0, 1, 2, 3])
df1
```

### Napravićemo jos jedan datafrejm koji će imati iste kolone ali nove zapise (redove). Nakon toga ćemo ih spojiti preko funkcije `concat` koja datafrejmove (najčesće sa istim kolonama) spaja tako sto združuje sve njihove redove. Funkcija prima listu datafrejmova koje je potrebno spojiti, ali i par opcionih parametara. Podrazumevano je da ova funkcija vrši konkatenaciju redova, međutim postavljanjem parametra `axis` na 1 naglašavamo da želimo da se združuju kolone.

```python
df2 = pd.DataFrame({'grad': ['SU', 'VR', 'CA', 'ZR'],
                    'glavni': ['NE', 'NE', 'NE', 'NE'],
                    'popul': ['150k', '<100k', '100k', '100k']},
                   index=[4, 5, 6, 7])
df12 = pd.concat([df1,df2])
df12
```

### Još jedna funkcija za spajanje datafrejmova je `merge`, koja na osnovu zajedičke kolone datafrejmova vrši različite vrste spajanja u zavisnosti od parametra `how`. U nasem primeru vršimo tzv. unutrašnje spajanje (`inner`) datafrejmova df1 i df3 na osnovu zajedničke kolone grad. 

```python
df3 = pd.DataFrame({'grad': ['BG', 'NS', 'NI', 'KG'],
                    'postbr': [11000, 22000, 18000, 30000]})
df13 = pd.merge(df1,df3,how='inner',on='grad')
df13
```

### Funkcija `join` koja vrši spajanje dva ili više datafrejma na osnovu indeksa. U našem primeru, df1 ima i indeks 3 koji ne postoji u df4, stoga će se kao rezultat spoljnoj spajanja (`outer`) javiti nedostajuće (NaN) vrednosti kolona df4 u tom redu. Slična situacija se događa i za indeks 40.

```python
df4 =  pd.DataFrame({'postbr': [11000, 22000, 18000, 44000],
                     'region': ['Beograd', 'Vojvodina', 'Juzna Srbija', 'Istocna Srbija']},
                    index=[0, 1, 2, 40])
df1.join(df4, how='outer')
```

### Metoda `compare` se koristi za pronalazak razlika između vrednosti u datafrejmovima. Vraća samo one redove i kolone gde postoje razlike, dok kod istovetnih vrednosti rezultat je NaN. Kao rezultat prikazuje dve kolone za svaku zajedničku kolonu:
### self – vrednosti iz prvog DataFrame-a (df1).
### other – vrednosti iz drugog DataFrame-a (df2).

```python
df1 = pd.DataFrame({'A': [0, 5, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [1, 2, 4], 'B': [4, 7, 6]})

df1.compare(df2)
```

### Metoda `pivot` koristi se za restruktuiranje datafrejma. U našem primeru, originalni datafrejm koji ima kolone date, city i temp, restruktuiraćemo tako što će datumi (jedinstvene vrednosti) postati redovi (tačnije indeksi), gradovi će postati kolone, a temperature će postati vrednosti u okviru ovako formiranog datafrejma.

```python
orig = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02'],
    'City': ['Lisboa', 'Barcelona', 'Granada'],
    'Temp': [5, 10, 3],
}

df = pd.DataFrame(orig)

print(df.pivot(index='Date', columns='City', values='Temp'))
```
